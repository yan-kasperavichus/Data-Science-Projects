#### Проверка вычислений эффективности обогащения

Построим функцию для вычисления целевого признака rougher.output.recovery:

```python
def recovery(df):
    for i in range(len(df)):
        C = df['rougher.output.concentrate_au']
        F = df['rougher.input.feed_au']
        T = df['rougher.output.tail_au']
        df['Recovery_Calc'] = ((C * (F-T)) / (F * (C-T))) * 100
    print('MAE:', mean_absolute_error(df.dropna()['rougher.output.recovery'], df.dropna()['Recovery_Calc']))
    df.pop('Recovery_Calc')
```

```python
recovery(df_GR_train)
```

**Вывод:**
- Значение recovery при вычислении на формуле близко к нулю, что значит эффективность обогащения рассчитана правильно.

- Практически все значения имеют выбросы ближе к нулю, однако, для лучшей работы модели удалить и выбросы, которые находтся за пределами третьего и первого квартиля. Для этого построим функцию.

Данная функция удаляет слишком высокие и низкие аномальные значения

```python
def delit_blowouts(df, first_column, last_column):
    for column in df.loc[:, first_column : last_column]:
        quant_25 = int(df[column].quantile(0.25))
        quant_75 = int(df[column].quantile(0.90))
        first_part=quant_25-1.5*(quant_75-quant_25)
        second_part=quant_75+1.5*(quant_75-quant_25)
        print('Столбец:', df[column].name)
        print('Строк было:', len(df[column]))
        print('Осталось строк:', len(df[(df[column] >= first_part) & (df[column] <= second_part)]))
        print('Строк было удалено:', len(df[column]) - len(df[(df[column] >= first_part) & (df[column] <= second_part)]))
        print()
        df_after_cleaning = df[(df[column] >= first_part) & (df[column] <= second_part)]
    return df_after_cleaning 
```

## Модель

### Функция для вычисления sMAPE

Напишем функцию для вычисления итогового sMAPE:

```python
def final_smape(target, prediction):
    target_rough = target['rougher.output.recovery']
    prediction_rough = prediction[:, 0]
    target_final = target['final.output.recovery']
    prediction_final = prediction[:, 1]
    smape_rough = (1 / len(target)) * (((abs(target_rough - prediction_rough)) / ((abs(target_rough) + abs(prediction_rough)) / 2)) * 100).sum()
    smape_final = (1 / len(target)) * (((abs(target_final - prediction_final)) / ((abs(target_final) + abs(prediction_final)) / 2)) * 100).sum()
    final_smape = 0.25 * smape_rough + 0.75 * smape_final
    return final_smape
```
